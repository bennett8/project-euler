```python
def find_first_factor(n):
    for number in range(2,n // 2):
        if n % number == 0:
            return number
            break
    print(n)
    return "Prime"
```

```python
def solve_problem(input):
    while True:
        first_factor = find_first_factor(input)
        if first_factor == "Prime":
            break
        input = input // first_factor
```
