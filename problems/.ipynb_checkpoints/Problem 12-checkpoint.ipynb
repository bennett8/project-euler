The sequence of triangle numbers is generated by adding the natural numbers. So the 7th triangle number would be 1 + 2 + 3 + 4 + 5 + 6 + 7 = 28. The first ten terms would be:

1, 3, 6, 10, 15, 21, 28, 36, 45, 55, ...

Let us list the factors of the first seven triangle numbers:

     1: 1
     3: 1,3
     6: 1,2,3,6
    10: 1,2,5,10
    15: 1,3,5,15
    21: 1,3,7,21
    28: 1,2,4,7,14,28

We can see that 28 is the first triangle number to have over five divisors.

What is the value of the first triangle number to have over five hundred divisors?

```python
%%time
import math
goal = 500

def solve_problem(goal_divisors):
    i = 1
    number = 1
    count = 0
    # create next triangular number
    while count <= goal:
        i += 1
        number += i
        # creates all divisors up to square root based on knowledge that half of divisors are below and half above (note that
        # odd number of divisors if number is perfect square)
        partial_list_divisors = [j for j in range(1,math.floor(math.sqrt(number+1))) if number % j == 0]
        count = len(partial_list_divisors) * 2
        if number % math.floor(math.sqrt(number)) == 0:
            count -= 1
    return number

solve_problem(goal)


# count divisors
```
