<p>Pentagonal numbers are generated by the formula, P<sub><var>n</var></sub>=<var>n</var>(3<var>n</var>−1)/2. The first ten pentagonal numbers are:</p>
<p style="text-align:center;">1, 5, 12, 22, 35, 51, 70, 92, 117, 145, ...</p>
<p>It can be seen that P<sub>4</sub> + P<sub>7</sub> = 22 + 70 = 92 = P<sub>8</sub>. However, their difference, 70 − 22 = 48, is not pentagonal.</p>
<p>Find the pair of pentagonal numbers, P<sub><var>j</var></sub> and P<sub><var>k</var></sub>, for which their sum and difference are pentagonal and D = |P<sub><var>k</var></sub> − P<sub><var>j</var></sub>| is minimised; what is the value of D?</p>
</div>

```python
pentagonal = lambda x: int(x*(3*x-1)/2)
```

```python
def is_pentagonal(number): # From wikipedia
    if (((24 * number + 1)**.5 + 1) / 6) == int(((24 * number + 1)**.5 + 1) / 6):
        return True
    else:
        return False
```

```python
print(is_pentagonal(5)) #True
print(is_pentagonal(5.1)) #False
print(is_pentagonal(144)) #False
print(is_pentagonal(117)) #True
```

```python
def pentagonal_list_maker(maximum):
    output_list = [1]
    i = 2
    while output_list[-1] < maximum:
        output_list.append(int(i*(3*i-1)/2))
        i += 1
    return output_list
# pentagonal_list_maker(1216)
```

```python
def main():
    # for pentagonal number:
    pentagonal_list = [1, 5]
    i = 3
    while True:
#         print(i)
        temp_value = pentagonal(i)
#         print(temp_value)
        temp_list = [j for j in pentagonal_list if temp_value-j in pentagonal_list]
        for item in temp_list:
            if is_pentagonal(temp_value + item):
                return temp_value - item
                break
#         print(temp_list)
        pentagonal_list.append(temp_value)
        i += 1
```

```python
%%time
main()
```
