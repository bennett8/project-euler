<p>In the United Kingdom the currency is made up of pound (£) and pence (p). There are eight coins in general circulation:</p>
<blockquote>1p, 2p, 5p, 10p, 20p, 50p, £1 (100p), and £2 (200p).</blockquote>
<p>It is possible to make £2 in the following way:</p>
<blockquote>1×£1 + 1×50p + 2×20p + 1×5p + 1×2p + 3×1p</blockquote>
<p>How many different ways can £2 be made using any number of coins?</p>

```python
# Used this link to learn dynamic programming: https://www.youtube.com/watch?v=DJ4a7cmjZY0
```

```python
def count(coins, coin_index, amount):
    if amount == 0:
        return 1
    if amount < 0:
        return 0
    if coin_index <= 0 and amount > 0:
        return 0
    
    return count(coins, coin_index, amount - coins[coin_index - 1]) + count(coins, coin_index - 1, amount)

coins_list = [1, 2, 5, 10, 20, 50, 100, 200]
coins_length = len(coins_list)
change = 200
count(coins_list, coins_length, change)
```

```python
#Threads 1 (I added print statements) - brute force (nested for loops - if use 0 200s, 0 100s, ... if 0 200s, 1 100, ...)
count = 1
num = 200

for a in range(2):
    num -= 200 * a
    save_num = num
    for b in range(save_num // 100 + 1):
        num = save_num - 100 * b
        save_num1 = num
        for c in range(save_num1 // 50 + 1):
            num = save_num1 - 50 * c
            save_num2 = num
            for d in range(save_num2 // 20 + 1):
                num = save_num2 - 20 * d
                save_num3 = num
                for e in range(save_num3 // 10 + 1):
                    num = save_num3 - 10 * e
                    save_num4 = num
                    for f in range(save_num4 // 5 + 1):
                        print('f: ' + str(f))
                        num = save_num4 - 5 * f
                        save_num5 = num
                        for g in range(save_num5 // 2 + 1):
                            num = save_num5 - 2 * g
                            save_num6 = num
                            print(save_num6)
                            count += 1

print(count)
```

```python
#Threads 2 - dynamic programming with explicit table (I think - convoluted with m and n)
import numpy as np
table=np.zeros([201,9])
table[0,1]=1
table[0,2]=2
table[0,3]=5
table[0,4]=10
table[0,5]=20
table[0,6]=50
table[0,7]=100
table[0,8]=200

for x in range(201):
    table[x,0]=x
    table[x,1]=1
    
for x in range(9):
    table[1,x]=1
    
for m in range(2,201):
    for n in range(2,9):
        if m/table[0,n]==1:
            table[m,n]=table[m,n-1]+1
        elif m/table[0,n]>1:
            a=int(m-table[0,n])
            table[m,n]=table[m,n-1]+table[a,n]
        else:
            table[m,n]=table[m,n-1]
print(table)
print(table[200,8])
```
