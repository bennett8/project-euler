```python
import math
```

```python
def n_choose_r(n,r):
    return math.factorial(n) // (math.factorial(n-r) * math.factorial(r))
```

```python
n_choose_r(23,10)
```

```python
%%time
solution = 0
for n in range(1,101):
    for r in range(1,n+1):
        if n_choose_r(n,r) > 1000000:
            solution +=1

print(solution)
```
