```python
# For each row
i = len(matrix) - 2
while i >= 0:
    # For each cell in row:
    j = len(matrix) - 2
    while j >= 0:
#         print(str(i) + str(j))
#         cell += min of cell to right and cell below
        matrix[i][j] += min(matrix[i][j+1], matrix[i+1][j])
        j -= 1
    i -= 1
```
